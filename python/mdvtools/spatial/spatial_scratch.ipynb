```python
# Let's also examine the specific transformation details for non-identity transforms
print("\n=== DETAILED TRANSFORMATION ANALYSIS ===\n")

def analyze_transform(name, transform):
    print(f"{name}:")
    print(f"  Type: {type(transform).__name__}")
    print(f"  Transform: {transform}")
    
    # Check if it's a sequence of transforms
    if isinstance(transform, sd.transformations.Sequence):
        print(f"  Sequence components:")
        try:
            transforms = getattr(transform, 'transforms', None)
            if transforms is not None:
                for i, comp in enumerate(transforms):
                    print(f"    {i}: {type(comp).__name__} = {comp}")
        except:
            print("    Could not access sequence components")
    
    # Check for scaling
    if isinstance(transform, sd.transformations.Scale):
        try:
            scale = getattr(transform, 'scale', None)
            if scale is not None:
                print(f"  Scale factors: {scale}")
        except:
            print("  Could not access scale factors")
    elif isinstance(transform, sd.transformations.Affine):
        try:
            affine = getattr(transform, 'affine', None)
            if affine is not None:
                print(f"  Affine matrix shape: {affine.shape}")
                print(f"  Affine matrix:\n{affine}")
        except:
            print("  Could not access affine matrix")
    elif isinstance(transform, sd.transformations.Translation):
        try:
            translation = getattr(transform, 'translation', None)
            if translation is not None:
                print(f"  Translation: {translation}")
        except:
            print("  Could not access translation")
    
    print()

# Check all elements for non-identity transforms (excluding tables which are AnnData)
all_elements = []
all_elements.extend([("images", name, img) for name, img in sdata.images.items()])
all_elements.extend([("labels", name, lbl) for name, lbl in sdata.labels.items()])
all_elements.extend([("points", name, pts) for name, pts in sdata.points.items()])

polygons = getattr(sdata, 'polygons', None)
if polygons is not None:
    all_elements.extend([("polygons", name, poly) for name, poly in polygons.items()])

# Skip tables as they are AnnData objects and don't have spatial transformations
print("Skipping tables (AnnData objects don't have spatial transformations)")

for category, name, element in all_elements:
    try:
        transform = get_transformation(element)
        if not isinstance(transform, sd.transformations.Identity):
            analyze_transform(f"{category}.{name}", transform)
    except Exception as e:
        print(f"Error processing {category}.{name}: {e}")
```

```python
# Let's check if the image metadata contains the Pixels.PhysicalSizeX information
print("=== CHECKING IMAGE METADATA FOR PIXEL SIZE ===\n")

# The React code expects metadata.Pixels.PhysicalSizeX to contain the pixel size
# This is what gets used in the calculation: Pixels.PhysicalSizeX / regionScale

morphology_image = sdata.images["morphology_focus"]
print(f"Checking image metadata for Pixels.PhysicalSizeX...")

# Check each scale level for OME-NGFF metadata that might contain pixel size
for scale_name in morphology_image.keys():
    scale_data = morphology_image[scale_name]
    print(f"\n--- {scale_name} ---")
    
    if hasattr(scale_data, 'attrs'):
        attrs = scale_data.attrs
        print(f"  All attrs keys: {list(attrs.keys())}")
        
        # Look for OME-NGFF multiscales metadata
        if 'multiscales' in attrs:
            multiscales = attrs['multiscales']
            print(f"  multiscales: {multiscales}")
            
            # Check if multiscales contains pixel size info
            if isinstance(multiscales, list) and len(multiscales) > 0:
                for i, scale_info in enumerate(multiscales):
                    print(f"    Scale {i}: {scale_info}")
                    if 'datasets' in scale_info:
                        for j, dataset in enumerate(scale_info['datasets']):
                            print(f"      Dataset {j}: {dataset}")
                            if 'coordinateTransformations' in dataset:
                                for k, transform in enumerate(dataset['coordinateTransformations']):
                                    print(f"        Transform {k}: {transform}")
                                    if transform.get('type') == 'scale':
                                        scale_values = transform.get('scale', [])
                                        print(f"          Scale values: {scale_values}")
                                        if len(scale_values) >= 2:
                                            # Usually the first two values are x and y pixel sizes
                                            pixel_size_x = scale_values[0]
                                            pixel_size_y = scale_values[1]
                                            print(f"          Pixel size X: {pixel_size_x}")
                                            print(f"          Pixel size Y: {pixel_size_y}")
        
        # Look for other OME-NGFF metadata that might contain pixel size
        for key, value in attrs.items():
            if any(term in key.lower() for term in ['pixel', 'physical', 'size', 'resolution', 'spacing']):
                print(f"  {key}: {value}")

print(f"\nCONCLUSION:")
print(f"  - The React code expects metadata.Pixels.PhysicalSizeX")
print(f"  - This should contain the actual pixel size (e.g., 0.2125 for Xenium)")
print(f"  - The calculation is: Pixels.PhysicalSizeX / regionScale")
print(f"  - If Pixels.PhysicalSizeX is missing, it returns 1, which explains the scaling issue")
print(f"  - We need to ensure the image metadata contains the correct pixel size information")
```

```python
# Let's create a function to extract pixel size from spatialdata and show how to fix the conversion
print("=== EXTRACTING PIXEL SIZE FROM SPATIALDATA ===\n")

def extract_pixel_size_from_spatialdata(sdata, image_name="morphology_focus"):
    """
    Extract pixel size from spatialdata OME-NGFF metadata.
    Returns the pixel size in micrometers, or None if not found.
    """
    if image_name not in sdata.images:
        print(f"Image '{image_name}' not found in spatialdata")
        return None
    
    image = sdata.images[image_name]
    
    # Check each scale level for OME-NGFF metadata
    for scale_name in image.keys():
        scale_data = image[scale_name]
        
        if hasattr(scale_data, 'attrs') and 'multiscales' in scale_data.attrs:
            multiscales = scale_data.attrs['multiscales']
            
            if isinstance(multiscales, list) and len(multiscales) > 0:
                for scale_info in multiscales:
                    if 'datasets' in scale_info:
                        for dataset in scale_info['datasets']:
                            if 'coordinateTransformations' in dataset:
                                for transform in dataset['coordinateTransformations']:
                                    if transform.get('type') == 'scale':
                                        scale_values = transform.get('scale', [])
                                        if len(scale_values) >= 2:
                                            # Usually the first two values are x and y pixel sizes
                                            pixel_size_x = scale_values[0]
                                            pixel_size_y = scale_values[1]
                                            print(f"Found pixel size: X={pixel_size_x}, Y={pixel_size_y}")
                                            return pixel_size_x  # Return X pixel size
    
    print("No pixel size found in OME-NGFF metadata")
    return None

# Try to extract pixel size from your spatialdata
pixel_size = extract_pixel_size_from_spatialdata(sdata)
print(f"\nExtracted pixel size: {pixel_size}")

if pixel_size is not None:
    print(f"\nSOLUTION:")
    print(f"  - Extract pixel size from spatialdata: {pixel_size}")
    print(f"  - Set region scale to this value instead of 1.0")
    print(f"  - This will make the React calculation work:")
    print(f"    Pixels.PhysicalSizeX / regionScale = 1 / {pixel_size} = {1/pixel_size:.4f}")
    print(f"    Then: scale / 0.2125 = {1/pixel_size:.4f} / 0.2125 = {(1/pixel_size)/0.2125:.4f}")
else:
    print(f"\nFALLBACK:")
    print(f"  - Use typical Xenium pixel size: 0.2125")
    print(f"  - Set region scale to 0.2125 instead of 1.0")
    print(f"  - This will make the React calculation work:")
    print(f"    Pixels.PhysicalSizeX / regionScale = 1 / 0.2125 = 4.7059")
    print(f"    Then: scale / 0.2125 = 4.7059 / 0.2125 = 22.15")

print(f"\nMODIFICATION NEEDED IN spatial_conversion.py:")
print(f"  - In _set_sdata_image function, change:")
print(f"    'scale': 1.0  # todo: read appropriate scale/roi from the image metadata")
print(f"  - To:")
print(f"    'scale': {pixel_size if pixel_size is not None else 0.2125}  # extracted from spatialdata")
```
