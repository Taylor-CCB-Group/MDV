```python
# setting up and serving the MDV project
base = os.path.expanduser('~/mdv')
project_path = os.path.join(base, 'taurus') # defining the location where the project metadata will be stored
p = MDVProject(os.path.expanduser(project_path), delete_existing=True)

# adding the two data sources to the project
p.add_datasource("B cells metadata 1", df_bcell)
p.add_datasource("B cells metadata 2", bcell_umap_gene_short)
p.add_datasource("B cells expression CD", bcell_gene_CD)
p.add_datasource("B cells expression UC", bcell_gene_UC)
```

```python
p.add_datasource("B cells expression proportion CD", bcell_CD_proportion)
p.add_datasource("B cells expression proportion UC", bcell_UC_proportion)
```

```python
p.add_datasource("B cells metadata 3", bcell_abundance)
```

```python
###Â adding an empty dataframe for the Landing page

empty = pd.DataFrame({'Summary' : []})

#empty

p.add_datasource("Summary", empty)
```

```python
# creating a scatter plot using the ScatterPlot implementation to show the 2 PCA clustering components
scatter2D_plot_bcells = ScatterPlot(
    title="B cells",
    params=["UMAP_1", "UMAP_2"],
    size=[400, 400],
    position=[10, 10]
)

# configuring the scatter plot
scatter2D_plot_bcells.set_default_color("#377eb8")
scatter2D_plot_bcells.set_brush("poly")
scatter2D_plot_bcells.set_opacity(0.8)
scatter2D_plot_bcells.set_radius(2)
scatter2D_plot_bcells.set_color_legend(display=True, position=[50, 5])
scatter2D_plot_bcells.set_axis_properties("x", {"label": "UMAP 1", "textSize": 13, "tickfont": 10})
scatter2D_plot_bcells.set_axis_properties("y", {"label": "UMAP 2", "textSize": 13, "tickfont": 10})
scatter2D_plot_bcells.set_color_by("Cell states")
```

```python
# creating a scatter plot using the ScatterPlot implementation to show the 2 PCA clustering components
scatter2D_plot_bcells_genes = ScatterPlot(
    title="B cells",
    params=["UMAP_1", "UMAP_2"],
    size=[400, 400],
    position=[10, 10]
)

# configuring the scatter plot
scatter2D_plot_bcells_genes.set_default_color("#377eb8")
scatter2D_plot_bcells_genes.set_brush("poly")
scatter2D_plot_bcells_genes.set_opacity(0.8)
scatter2D_plot_bcells_genes.set_radius(2)
scatter2D_plot_bcells_genes.set_color_legend(display=True, position=[50, 5])
scatter2D_plot_bcells_genes.set_axis_properties("x", {"label": "UMAP 1", "textSize": 13, "tickfont": 10})
scatter2D_plot_bcells_genes.set_axis_properties("y", {"label": "UMAP 2", "textSize": 13, "tickfont": 10})
scatter2D_plot_bcells_genes.set_color_by("HSPE1")
```
